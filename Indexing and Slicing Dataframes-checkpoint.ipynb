```python
# You can access the dataframes in two ways, label based indexing
# or position based indexing 
# There are index labels(COLUMN)(0,1,2,3,4,5 in this case) and 
# column labels(ROW) (Address,City,Country,Employees,ID,Name,State)
# set_index only shows the updated table and prints it, it doesn't store
# The changes in the data frame
df7.set_index("Address")
```

```python
# Our dataframe doesn't change even though we used set_index above
df7
```

```python
# For storing updation we store the updated dataframe in a new variable 
# Or the Same variable 
df7 = df7.set_index("Address")
```

```python
# We can see that the table has been updated and stored
df7
```
